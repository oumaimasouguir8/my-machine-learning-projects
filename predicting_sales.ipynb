```python
# Apply log transformation
train_df['YrSold'] = np.log(train_df['SalePrice'])
```

```python
# Convert categorical variable into dummy
train_df = pd.get_dummies(train_df)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train_df[['SalePrice','YrSold']])
y = np.asanyarray(train_df[['YearBuilt']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test_df[['MoSold','YrSold']])
x = np.asanyarray(test_df[['MiscVal','MSSubClass']])
y = np.asanyarray(test_df[['GrLivArea']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
